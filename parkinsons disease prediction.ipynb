# Data Preprocessing

```python
data.drop(['name'], axis=1, inplace=True)
X = data.drop('status', axis=1)
Y = data['status']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=40)
```

# Model Evaluation Function

```python
def evaluate_model(model, X_train, Y_train, X_test, Y_test):
    model.fit(X_train, Y_train)
    train_preds = model.predict(X_train)
    test_preds = model.predict(X_test)
    train_accuracy = accuracy_score(Y_train, train_preds)
    test_accuracy = accuracy_score(Y_test, test_preds)
    train_conf_matrix = confusion_matrix(Y_train, train_preds)
    test_conf_matrix = confusion_matrix(Y_test, test_preds)
    train_classification_report = classification_report(Y_train, train_preds)
    test_classification_report = classification_report(Y_test, test_preds)
    return train_accuracy, test_accuracy, train_conf_matrix, test_conf_matrix, train_classification_report, test_classification_report
```

# Models

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Naive Bayes': GaussianNB(),
    'K Nearest Neighbors': KNeighborsClassifier(weights='distance'),
    'Support Vector Machine': SVC(kernel='linear')
}
```

# Evaluate and print results

```python
for name, model in models.items():
    train_accuracy, test_accuracy, train_conf_matrix, test_conf_matrix, train_report, test_report = evaluate_model(model, X_train, Y_train, X_test, Y_test)
    print(f"Model: {name}")
    print(f"Train Accuracy: {train_accuracy}")
    print(f"Test Accuracy: {test_accuracy}")
    print(f"Confusion Matrix (Train):\n{train_conf_matrix}")
    print(f"Confusion Matrix (Test):\n{test_conf_matrix}")
    print(f"Classification Report (Train):\n{train_report}")
    print(f"Classification Report (Test):\n{test_report}")
    print('-' * 50)
```

# Save SVM model

```python
svm_model = SVC(kernel='linear').fit(X_train, Y_train)
with open('deploy_SVM.pkl', 'wb') as file:
    pickle.dump(svm_model, file)
```

# K-fold Cross-Validation:

```python
from sklearn.model_selection import cross_val_score


scores = cross_val_score(svm_model, X_train, Y_train, cv=5)

print("Cross-validation scores:", scores)
print("Mean accuracy:", scores.mean())
```

# Handling Class Imbalance:

```python
class_weights = dict(zip([0, 1], ((len(Y_train) - Y_train.sum()) / len(Y_train), Y_train.sum() / len(Y_train))))
svm = SVC(kernel='linear', class_weight=class_weights)
svm.fit(X_train, Y_train)
```
